```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# AULA DE MATPLOTLIB - PARTE 1
x = np.linspace(0, 5, 11)
x
```

```python
y = x * x
y
```

```python
# Funcional
plt.plot(x, y, color='r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
#plt.subplots(nrows=1, ncols=2)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes.plot(x, y)
axes.set_xlabel('Eixo X')
axes.set_title('Título')
axes2.plot(y, x)
```

```python
# AULA DE MATPLOTLIB - PARTE 2
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Teste')
fig
```

```python
fig, axes = plt.subplots(figsize=(8, 10))
axes.plot(x, y, 'r-', label='x ^ 2')
axes.plot(y, x, 'g--', label='Verde Tracejado')
axes.legend(loc=0)
```
